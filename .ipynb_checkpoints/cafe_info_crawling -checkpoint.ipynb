```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import time
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
driver.get("https://map.naver.com/v5/search/줄라이모닝")

# 팝업 창 제거
# driver.find_element(By.CSS_SELECTOR, "button#intro_popup_close").click()

# 검색창에 텍스트 입력
wait = WebDriverWait(driver, 10)  # 최대 10초 동안 대기
search_box = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "input.input_search")))

# search_box.send_keys("마라탕")

# 3초 대기 (옵션)
time.sleep(3)

# 검색 실행
search_box.send_keys(Keys.ENTER)
```

```python
# 크롤링
for p in range(20):
    # 5초 delay
    time.sleep(2)
    
    js_script = "document.querySelector(\"body > app > layout > div > div.container > div.router-output > "\
                "shrinkable-layout > search-layout > search-list > search-list-contents > perfect-scrollbar\").innerHTML"
    raw = driver.execute_script("return " + js_script)

    html = BeautifulSoup(raw, "html.parser")

    contents = html.select("div > div.ps-content > div > div > div .item_search")
    for s in contents:
        search_box_html = s.select_one(".search_box")

        name = search_box_html.select_one(".title_box .search_title .search_title_text").text
        print("식당명: " + name)
        try:
            phone = search_box_html.select_one(".search_text_box .phone").text
        except:
            phone = "NULL"
        print("전화번호: " + phone)
        address = search_box_html.select_one(".ng-star-inserted .address").text
        print("주소: " + address)

        print("--"*30)
    # 다음 페이지로 이동
    try:
        next_btn = driver.find_element_by_css_selector("button.btn_next")
        next_btn.click()
    except:
        print("데이터 수집 완료")
        break

# 크롭 웹페이지를 닫음
driver.close()
```

```python
# Selenium을 사용하여 웹 드라이버 초기화
driver = webdriver.Chrome()
driver.get("https://map.naver.com/p/search/카페 네쥬")

# 필요한 요소가 로드될 때까지 대기
wait = WebDriverWait(driver, 5)  # 최대 5초 동안 대기
element_present = EC.presence_of_element_located((By.ID, "searchIframe"))
wait.until(element_present)

driver.switch_to.frame("searchIframe")

# ActionChains 객체 생성
# actions = ActionChains(driver)
# iframe = driver.find_element(By.ID, "searchIframe")
# actions.context_click(iframe).perform()
# actions.click(iframe).perform()

# BeautifulSoup 객체를 생성합니다.
html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')

current_url = driver.current_url
url_point = current_url.split("/")
print(url_point)

if len(url_point) > 6:
    print("상세 페이지")
    
    # entryIframe 으로 전환
    driver.switch_to.default_content()
    element_present = EC.presence_of_element_located((By.ID, "entryIframe"))
    wait.until(element_present)
    driver.switch_to.frame("entryIframe")
    
    
else:
    print("목록 페이지")
    # 주소 값에 "대구" or "경산" 지역의 카페만 선택
    span_tags = soup.find_all('span', class_='Pb4bU')
    print(span_tags)

    # span 태그의 텍스트를 가져옵니다.
    if span_tags:
        for span_element in span_tags:
            if "대구" in span_element.text:
                print(span_element)
                # BeautifulSoup 객체에서 찾은 span_element의 텍스트를 바탕으로 Selenium을 이용해 같은 요소를 찾습니다.
                selenium_span_element = driver.find_element(By.XPATH, f"//span[contains(text(), '{span_element.text}')]")
                # Selenium을 이용해 span_element의 부모 div 요소를 찾습니다.
                parent_div = selenium_span_element.find_element(By.XPATH, "./ancestor::div[contains(@class, 'ouxiq icT4K')]")

                break
        
        a_tag_to_click = parent_div.find_element(By.XPATH, ".//a[contains(@class, 'P7gyV')]")
        a_tag_to_click.click()
        
        # entryIframe 으로 전환
        driver.switch_to.default_content()
        element_present = EC.presence_of_element_located((By.ID, "entryIframe"))
        wait.until(element_present)
        driver.switch_to.frame("entryIframe")
        
        html_source = driver.page_source
        soup = BeautifulSoup(html_source, 'html.parser')
        
        # 카페 정보 추출q
        cafe_name = soup.find('span', class_='Fc1rA').text
        cafe_type = soup.find('span', class_='DJJvD').text
        reviews_div = soup.find('div', class_='dAsGb')
        starring = float(reviews_div.find('span', class_='PXMot LXIwF').text.strip()[2:]) if reviews_div.find('span', class_='PXMot LXIwF') else 0
        print(reviews_div.find_all('span', class_='PXMot'))
        visitor_reviews = int(reviews_div.find_all('span', class_='PXMot')[0].text.strip().split(" ")[1])
        blog_reviews = int(reviews_div.find_all('span', class_='PXMot')[1].text.strip().split(" ")[1])
        
        
        df.loc[len(df)] = [cafe_name, cafe_type, starring, visitor_reviews, blog_reviews, "", "", "", "", "", "", ""]
        print(df.loc[len(df)-1])
    else:
        print("해당 태그를 찾을 수 없습니다.")
    

# BeautifulSoup 객체 생성
# 페이지 소스 가져오기
# driver.switch_to.frame(iframe)
# html_source = driver.page_source
# soup = BeautifulSoup(html_source, 'html.parser')

# div_tag = soup.find('div', class_='Ryr1F')
# print(div_tag)

# # div 태그 존재 여부 확인
# if div_tag:
#     print("div 태그가 존재합니다.")
# else:
#     print("div 태그가 존재하지 않습니다.")

# # 웹 드라이버 종료
# driver.close()
```
