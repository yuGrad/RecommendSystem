```python
import pandas as pd
import numpy as np
```

```python
cafe_df = pd.read_csv("./cafe_data/cafe_data_v1.csv",  encoding="utf-8")
```

```python
cafe_df.columns
```

```python
cafe_df.info()
```

```python
cafe_df['키워드'] = cafe_df['키워드'].apply(str)
```

```python
# 키워드 {] 문자열을 딕셔너리로 변환
import ast

cafe_df['키워드'] = cafe_df['키워드'].apply(ast.literal_eval)
```

```python
cafe_df['키워드']
```

```python
# 각 카페 행의 키워드 컬럼에서 각각의 키워드를 추출해 리스트 -> 모든 키워드 추출
keywords_map = {}

for idx, row in cafe_df.iterrows():
    # Extract the keyword string from the row
    if row['키워드'] == {} or row['키워드'] == "{}":
        continue

    try:
        keywords_dict = dict(row['키워드'])
    except:
        keywords_dict = ast.literal_eval(row['키워드'])
        row['키워드'] = ast.literal_eval(row['키워드'])
        cafe_df.iloc[idx] = row
    
    keywords_set = {keyword for keyword in keywords_dict.keys()}
        
    for keyword in keywords_set:
        if keyword in keywords_map:
            keywords_map[keyword] = keywords_map[keyword] + 1
        else:
            keywords_map[keyword] = 1
```

```python
keywords_map
```

```python
unique_keywords = [keyword for keyword in keywords_map if keywords_map[keyword] > 5]
```

```python
categories = {
    # 음료, 음식
    'dessert': ['디저트가 맛있어요', '빵이 맛있어요', '음식이 맛있어요'],       
    'various_menu': ['메뉴 구성이 알차요', '종류가 다양해요'],
    'special_menu': ['특별한 메뉴가 있어요', '특색 있는 제품이 많아요'],

    # 시설
    'large_store': ['매장이 넓어요', '룸이 잘 되어있어요'],
    'background': ['야외 공간이 멋져요', '뷰가 좋아요'],
    
    # 분위기
    'comfort&talking': ['분위기가 편안해요', '대화하기 좋아요', '아늑해요', '집중하기 좋아요', '차분한 분위기에요',  '오래 머무르기 좋아요'],
    'trendy&unique': ['사진이 잘 나와요', '음악이 좋아요', '인테리어가 멋져요', '컨셉이 독특해요'],

    # 가격대 
    'low_price': ['가격이 합리적이에요', '가성비가 좋아요', '양이 많아요'],
    'high_price': ['특별한 날 가기 좋아요', '비싼 만큼 가치있어요'], 

    # 기타
    'gift_packaging': ['선물하기 좋아요', '포장이 깔끔해요', '주문제작을 잘해줘요'],
    'parking': ['주차하기 편해요'],
    
    # 공통 키워드 부분
    'beverage': ['차가 맛있어요', '음료가 맛있어요', '커피가 맛있어요'],
    'clean_store': ['화장실이 깨끗해요', '시설이 깔끔해요', '매장이 청결해요'],
    'service': ['좌석이 편해요', '친절해요',  '음식이 빨리 나와요'],
}
```

```python
# 키워드를 카테고리로 매핑하는 함수
def map_keywords_to_categories(keyword_dict):
    category_dict = {cat: 0 for cat in categories.keys()}
    for keyword, score in keyword_dict.items():
        for category, keywords in categories.items():
            if keyword in keywords:
                category_dict[category] += int(score)
    return category_dict
```

```python
# 각 카테고리별 점수를 계산
cafe_df['분류별 점수'] = cafe_df['키워드'].apply(map_keywords_to_categories)

# 데이터프레임에 각 카테고리별 점수를 새 열로 추가
for category in categories.keys():
    cafe_df[category] = cafe_df['분류별 점수'].apply(lambda x: x[category])

# 필요 없는 "분류별 점수" 열을 삭제
cafe_df = cafe_df.drop(columns=['분류별 점수'])
```
