```python
import pandas as pd
import numpy as np
```

```python
daegu_df = pd.read_csv("./daegu_cafe.csv",  encoding="utf-8")
gyungsan_df = pd.read_csv("./gyungsan_cafe.csv",  encoding="utf-8")
```

```python
gyungsan_df.isnull().sum()
```

```python
# 두 지역 데이터프레임 합치기
init_df = pd.concat([daegu_df, gyungsan_df])
```

```python
len(init_df)
```

```python
init_df.isnull().sum()
```

```python
init_df = init_df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"], axis=1)
```

```python
init_df
```

```python
# 주소 컬럼이 결측치인 행 삭제
init_df = init_df.dropna(subset=["주소"])
```

```python
init_df.describe()
```

```python
# 주소가 초기값인 카페는 삭제
init_df = init_df[init_df['주소'] != "초기값"]
```

```python
# "초기값", "NO INFO"와 같은 결측치를 0 or 빈 문자열로 대체
init_df[['별점', '방문자리뷰수', '블로그리뷰수']] = init_df[['별점', '방문자리뷰수', '블로그리뷰수']].replace(["초기값", "검색결과 없음", "NO INFO"], "0")
init_df["메뉴"] = init_df["메뉴"].replace(["공사중"], "NO INFO")
init_df["키워드"] = init_df["키워드"].replace(["리뷰 수 부족"], "{}")
```

```python
init_df.describe()
```

```python
# 각 컬럼 타입 확인 후 컬럽 타입 변환
init_df.info()
init_df['별점'] = pd.to_numeric(init_df['별점'], errors="coerce")
init_df['방문자리뷰수'] = pd.to_numeric(init_df['방문자리뷰수'], errors="coerce")
init_df['블로그리뷰수'] = pd.to_numeric(init_df['블로그리뷰수'], errors="coerce")
```

```python
# 사업장명 컬럼 드랍
init_df = init_df.drop(columns="사업장명", axis=1)
```

```python
# 주소가 동일한 카페 확인 후 드랍
init_df[init_df.duplicated(["전화번호", "주소"], keep="last")]
init_df = init_df.drop_duplicates(["전화번호", "주소"], keep="last")
```

```python
init_df = init_df.reset_index().drop(columns="index", axis=1)
```
